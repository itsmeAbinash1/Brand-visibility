```python
#FETCHING THE DATA FROM THE API FOR EACH KEYWORD AND PRINTING THE RESULTS.
import requests
import pandas as pd
import time

KEYWORDS = [
    "Smart Phone", "Washing Machine", "laptop", "headphones",
    "smartwatch", "tablet", "gaming", "camera",
    "wireless earbuds", "Bluetooth", "camera"
]

API_KEY = "4d6b57a69cfd907adc136e4ce8558e26d973ca538786473d2f685f299a56e641"
url = "https://serpapi.com/search.json"

all_products = []

for keyword in KEYWORDS:
    print(f"\n Searching: {keyword}")
    keyword_count = 0
    start = 0

    while keyword_count < 100:
        params = {
            "engine": "google_shopping",
            "q": keyword,
            "location": "India",
            "api_key": API_KEY,
            "start": start,
            "num": 40
        }

        response = requests.get(url, params=params)
        data = response.json()

        results = data.get("shopping_results", [])

        if not results:
            print(f"No more results at start={start}")
            break

        for item in results:
            all_products.append({
                "keyword": keyword,
                "title": item.get("title"),
                "price": item.get("extracted_price"),
                "raw_price": item.get("price"),
                "rating": item.get("rating"),
                "reviews": item.get("reviews"),
                "platform": item.get("source"),
                "position": item.get("position"),
                "delivery": item.get("delivery"),
                "link": item.get("link"),
                "thumbnail": item.get("thumbnail")
            })

        keyword_count += len(results)
        start += len(results)
        print(f"   Page fetched | start={start - len(results)} | Got {len(results)} results | Total: {keyword_count}")

        time.sleep(1)  # small delay to avoid rate limiting

    print(f" {keyword}: {keyword_count} results collected")

df = pd.DataFrame(all_products)
print(f"\n{'='*50}")
print(f"Total results collected: {len(df)}")
print(f"Results per keyword:")
print(df.groupby("keyword").size())
```

```python
#Q3: Is there a relationship between rating and position?
final_dataset_df["rating"] = pd.to_numeric(
    final_dataset_df["rating"],
    errors="coerce"
)

final_dataset_df["position"] = pd.to_numeric(
    final_dataset_df["position"],
    errors="coerce"
)

corr = final_dataset_df["rating"].corr(
    final_dataset_df["position"]
)
#scatterplot

print("Correlation:", round(corr, 3))

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,3))

sns.scatterplot(
    data=final_dataset_df,
    x="rating",
    y="position"
)

plt.title("Rating vs Position")
plt.xlabel("Rating")
plt.ylabel("Position")

plt.show()
```

```python
#Q4: Do products with higher reviews rank better?
final_dataset_df["reviews"] = pd.to_numeric(
    final_dataset_df["reviews"],
    errors="coerce"
)

final_dataset_df["position"] = pd.to_numeric(
    final_dataset_df["position"],
    errors="coerce"
)

corr = final_dataset_df["reviews"].corr(
    final_dataset_df["position"]
)

print("Correlation:", round(corr,3))

#scatterplot
plt.figure(figsize=(6,4))

plt.scatter(
    final_dataset_df["reviews"],
    final_dataset_df["position"]
)

plt.xscale("log")

plt.title("Reviews vs Position (Log Scale)")
plt.xlabel("Reviews")
plt.ylabel("Position")

plt.show()
```

```python
#Q5: Which combination of factors (price, rating, reviews) influences top-ranking products?
final_dataset_df["rank_group"] = pd.cut(
    final_dataset_df["position"],
    bins=[0, 10, 30, 100],
    labels=["Top 10", "Top 30", "Below 30"]
)
review_analysis = (
    final_dataset_df
    .groupby("rank_group")["reviews"]
    .mean()
    .round(0)
)
print(review_analysis)

review_analysis.plot(
    kind="bar",
    figsize=(6,3)
)

plt.title("Average Price, Rating and Reviews by Ranking Group")
plt.ylabel("Average Value")
plt.xticks(rotation=0)

plt.show()
```
